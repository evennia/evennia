```python
import requests
```

```python
url="192.168.68.95:11434/api/chat"
```

```python
prompt = [{
  "model": "llama3",
  "messages": [
    {
      "role": "user",
      "content": "why is the sky blue?"
    }
  ]
}]
```

```python
response = requests.get(url, prompt)
response_json = response.json()
for i in response_json:
    print (i, "\n")
```
